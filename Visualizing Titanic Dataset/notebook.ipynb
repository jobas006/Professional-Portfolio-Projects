```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import csv
```

```python
# Loading the “titanic_data.csv” file
# Converting it to a DataFrame

df = pd.read_csv('titanic_data.csv', sep=',',header=None)

titanticData = df.rename(columns=df.iloc[0]).drop(0)
titanticData
```

```python
# Changing the datatypes of "SibSp" and "Parch" to an 'int' for future computations

titanticData['SibSp']= titanticData.SibSp.astype(int)
titanticData['Parch']= titanticData.Parch.astype(int)

titanticData.dtypes
```

```python
# Adding a new column called “FamilySize”
# Column is the result of columns SibSp  -Parch

titanticData['FamilySize'] = titanticData.SibSp - titanticData.Parch

titanticData.head(14)
```

```python
# Removing unnecessary column and pointing the result to 
# A variable called 'newTitanticData'

newTitanticData = titanticData.drop(['Cabin','Name','Ticket', 'SibSp', 'Parch'], axis = 1)
newTitanticData
```

```python
#changing the datatypes of these two columns for future computations

#newTitanticData["Survived"] = newTitanticData.Survived.astype(object)

#newTitanticData["Sex"] = newTitanticData.Survived.astype('category')
newTitanticData["Age"] = newTitanticData.Age.astype(float)

#Replaceing empty values in Embarked with 'Other'

newTitanticData['Embarked'].fillna('Other', inplace = True)
newTitanticData
```

```python
# Creating a category system for both the Embarked and Pclass columns
# Using a dictionary

embarkedCategory = pd.Categorical(newTitanticData['Embarked'])
newTitanticData['Embarked'] = embarkedCategory.rename_categories({'S':'Southampton','C':'Cherbourg','Q':'Queenstown'})

pClassCategory = pd.Categorical(newTitanticData['Pclass'])
newTitanticData['Pclass'] = pClassCategory.rename_categories({'1':'Upper Class','2':'Middle Class','3':'Lower Class'})

survivedCategory = pd.Categorical(newTitanticData['Survived'])
newTitanticData['Survived'] = survivedCategory.rename_categories({'0': False, '1': True})

newTitanticData.head(9)
```

```python
# Mean of ages grouped by both Sex and Pclass
groupMean = newTitanticData.groupby(['Sex', 'Pclass'])['Age'].mean()
groupMean
```

```python
# Just checking some of the rows that have
# NaN values in "Age" column

ageFilter = newTitanticData['Age'].isnull()
newTitanticData[ageFilter]
```

```python
# Created a function where it when called, it will apply the mean of ages 

def remove_na_ages(row):

    if pd.isnull(row['Age']):
        return groupMean[row['Sex'],row['Pclass']]
    else:
        return row['Age']

newTitanticData['Age'] = newTitanticData.apply(remove_na_ages, axis=1)
newTitanticData.head(10)
```

```python
# Creates a new column called "AgeGroup" with labels that groups the 
# ages based of the labels

ageCategory= ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']


newTitanticData["AgeGroup"] = pd.cut(newTitanticData.Age, range(0, 81, 10), right=False,
       labels=ageCategory)

#Turns Age column into a int to remove decimal point

newTitanticData["Age"] = newTitanticData.Age.astype(float)
newTitanticData.head(20)
#print(upgradedTitantic['AgeGroup'].unique())
```

```python
#Show the number of survivors and non-survivors using “groupby”
numSurvivor = newTitanticData.groupby(['Survived'])['Age'].count()
numSurvivor 
```

```python
#Creating a crosstab table to show survivors and non-survivors for each gender
crossTable = pd.crosstab(newTitanticData['Survived'],newTitanticData['Sex'])
crossTable
```
